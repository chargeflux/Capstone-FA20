```python
import pandas as pd
import numpy as np
```

```python
extract_csv = pd.read_csv("Simulator.csv")
```

```python
from sklearn.tree import DecisionTreeClassifier

label = 'cycles'

tree = DecisionTreeClassifier(criterion="entropy")
tree.fit(extract_csv.drop([label], axis=1), extract_csv[label])
```

```python
newdf = pd.DataFrame({'imp':tree.feature_importances_, 'feat':extract_csv.drop(label, 1).columns.values})
newdf = newdf.sort_values(by = 'imp', axis = 0)
newdf
```

```python
important_columns = list(newdf[newdf['imp'] > 0]['feat'])
```

```python
important_columns.append('l1cache_val_32k')
important_columns.append('l2cache_val_2m')
important_columns.append('l2cache_val_1m')
important_columns.append('l3cache_val_32m')
important_columns.append('width_val_8')
```

```python
important_columns
```

```python
extract_csv_slim = extract_csv[important_columns]
extract_csv_slim
```

```python
extract_csv_slim.shape
```

```python
#Add on the label again
extract_csv_slim_full = extract_csv_slim
extract_csv_slim_full[label] = extract_csv[label]
extract_csv_slim_full
```

```python
extract_csv_slim_full['fluidanimate'] = extract_csv['fluidanimate'].values
```

```python
train_df = extract_csv_slim_full[extract_csv_slim_full['fluidanimate'] != 1.0]
test_df = extract_csv_slim_full[extract_csv_slim_full['fluidanimate'] == 1.0]
```

```python
X_Train = train_df.drop(label,1)
Y_Train = train_df[label]
X_Test = test_df.drop(label,1)
Y_Test = test_df[label]
```

```python
from sklearn.model_selection import GridSearchCV
from sklearn.model_selection import KFold
from sklearn.pipeline import Pipeline
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures
from sklearn.model_selection import cross_val_score
```

```python
lr = LinearRegression()
lr.get_params().keys()
```

```python
crossvalidation = KFold(n_splits=10)

for i in range(1,4):
    polynomial = PolynomialFeatures(degree=i)
    X_New = polynomial.fit_transform(X_Train)
    model = lr.fit(X_New, Y_Train)
    scores = cross_val_score(model, X_New, Y_Train, scoring="neg_root_mean_squared_error", cv=crossvalidation,
 n_jobs=1)
    print("Degree "+str(i)+" polynomial gave an RMSE: " + str(np.mean(np.abs(scores))))
```

```python
import numpy as np
from sklearn import linear_model
from sklearn.metrics import mean_squared_error
polynomial = PolynomialFeatures(degree=1)
#X_New = polynomial.fit_transform(X_Train)
linear_fitted = lr.fit(X_Train, Y_Train)
y_pred = linear_fitted.predict(X_Test)
mean_squared_error(Y_Test, y_pred, squared=False) #False for RMSE
```

# Ridge Regression (L2)

```python
ridge_model = linear_model.RidgeCV(alphas=np.logspace(-10, 10, 30))
ridge_fitted = ridge_model.fit(X_Train, Y_Train)
y_pred = ridge_fitted.predict(X_Test)
mean_squared_error(Y_Test, y_pred, squared=False) #False for RMSE
```
